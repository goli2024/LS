```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('diabetes.csv')
```

```python
X  = data.drop(columns=['Outcome'])
Y = data['Outcome']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,random_state = 42,test_size = 0.2)
```

```python
model = DecisionTreeClassifier(random_state  = 42)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
repot = classification_report(y_test,y_pred)
conf_matrix = confusion_matrix(y_test,y_pred)
```

```python
print("Acuuracy : ",accuracy)
print("\nClassification Report : ",repot)
print("\nConfusion matrix : ",conf_matrix)
```

```python
random_sample = X.sample(n=1,random_state=42)
random_sample_pred = model.predict(random_sample)
print("\nRandom Sample prediction : ","Diabetic" if random_sample_pred[0] == 1 else "Non-Diabetic")
```

```python
plt.figure(figsize=(30,15))
plot_tree(model,feature_names=X.columns,class_names=['Non-Diabetic','Diabetic'],filled = True,rounded=  True,fontsize = 14)
plt.title("Decision Tree for Diabetes prediction",fontsize = 16)
plt.show()
```
